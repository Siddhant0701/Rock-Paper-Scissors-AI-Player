```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
ys = []
xs = []
winrates = []
with open('collected.csv', 'r') as file:
    reader = csv.reader(file, delimiter=',')

    next(reader)

    for row in reader:
        data = row[2]
        ys.append([int(i) for i in data])
        xs.append(list(range(1, int(row[1])+1)))
        winrates.append(row[3])
```

```python
def get_wins(data):
    wins = []
    w=0
    for count,i in enumerate(data):
        if i == 'w':
            w += 1
        wins.append(w/(count+1))
    return wins
```

```python
wins = list(map(get_wins, winrates))
```

```python
y0, y1 = np.array(ys[0]), np.array(ys[1])
x0, x1 = np.array(xs[0]), np.array(xs[1])
```

```python
#normalize y0 and y1
y0 = (y0 - np.min(y0)) / (np.max(y0)-np.min(y0))
y1 = (y1 - np.min(y1)) / (np.max(y1)-np.min(y1))
```

```python
plt.plot(x0,y0)
plt.plot(x0, wins[0])
plt.show()
plt.plot(x1,y1)
plt.plot(x1, wins[1])
plt.show()
```
